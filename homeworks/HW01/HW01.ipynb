1 Импорт библиотеки numpy

```python
import numpy as np
```

2 Примеры создания массивов

```python
#Создание массиа из обычного списка
m1 = np.array([0, 1, 2, 3, 4, 5])
display(m1)

#Создание массива с использованием функций zeros, ones, arange
m2 = np.zeros(6)   #создание массива нулей
m3 = np.ones(6)    #создание массива едениц
m4 = np.arange(6)  #создание массива с диапозоном значений
display(m2, m3, m4)
```

3 Демонстрация атрибутов 

```python
m5 = np.arange(6).reshape(2, 3)
display(m5)
display(m5.shape)  #показывает размерность массива
display(m5.dtype)  #выводит информацию о типах данных массива
```

4 Примеры индексирования и срезов 

```python
x = m5[0,2]          #выборка конкретного элемента по индексу
display(x) 
srez = m5[1, 0:3]    #пример среза, а конкретней 2я строка и элементы с 0 по 2 включительно массва m5 
display(srez)
```
